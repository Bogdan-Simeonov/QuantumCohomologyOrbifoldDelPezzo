# Basic setup

We define variables qi which will give us monomials from which we infer partitions and subpartitions. The variables d,dd are the divisor variables.

We also define a variable that we want to solve for corresponding to $\langle \delta_i\rangle_{part}, \langle \delta_1, \delta_n, \delta\rangle_{part}$. Probably need a dictionary. Let's name the 1pt ones a and the 3pt ones b. We have that $a_1=1, b_0=\frac{1}{k},b_1=\frac{-1}{k^2}$

```sage
n=7
qnames      = ["q%d"%i for i in range(1,n+1)]
dnames      = ["d{}".format(i) for i in range(1, n+1)]       # d1,...,dn
ddnames     = ["dd{}".format(i) for i in range(1, n+1)]      # dp1,...,dpn

def b_name(int,part):
    if part == ():
        return "b_trivial"
    else:
        return "b_{}_{}".format(int,"".join(map(str,part)))

b_paramnames = []
for btotal in range(1,n+1):
    for part in Partitions(btotal):
        for int in range(1,n+1):
            b_paramnames.append(b_name(int,tuple(part)))
print(b_paramnames)

def a_name(part):
    if part == ():
        return "a_trivial"
    else:
        return "a_{}".format("".join(map(str,part)))
a_paramnames = []
for atotal in range(1,n+1):
    for part in Partitions(atotal):
        a_paramnames.append(a_name(tuple(part)))
print(a_paramnames)



# define the ring
R=PolynomialRing(QQ,qnames+dnames+ddnames+a_paramnames+b_paramnames)
gens = R.gens()
q=gens[0:n]
d=gens[n:2*n]
dd=gens[2*n:3*n]



#here's how to get the valuation
f = (q[0])^2+q[1]+q[2]
f
f.degree(q[1])


```

# Defining the RHS and LHS in WDVV

Now let's do WDVV for $\delta_1, \delta{n-1},\phi_D, \phi_{D'}$ in a class of degree $n/k$.

```sage
def kRHS_from_monomial(poly):
    kRHS=0
    # this is computing Ljcdot the class we are in+1, multiplied by k so that we are not in the fraction field
    A=poly.degree()-1+k[0]

    #this is encoding the common part of D' cdot the class. we need to also remove Dcdot L_j
    sum = 0
    for i in range(n):
        sum += R(poly).degree(q[i])*dd[i]


    for j in range(n):
        if R.monomial_divides(q[j],poly):
            ainv = a_name(partition_from_poly(R.monomial_quotient(poly,q[j])))
            kRHS+=d[j]*(S(A)-k[0]*poly.degree(q[j]))*(sum-dd[j])*S(ainv)
    return kRHS

kRHS_from_monomial(q[0]*q[1]*q[2])
```

This defines the intersection of $D$ with $\beta$ where $\beta$ is associated to a monomial in $q_i$. The result is some linear thing in the di.

```sage
def divisor_intersection_d(poly): 
    divisorsum=0
    for i in range(n):
        divisorsum+=d[i]*poly.degree(q[i])
    return divisorsum

def divisor_intersection_dd(poly): 
    ddivisorsum=0
    for i in range(n):
        ddivisorsum+=dd[i]*poly.degree(q[i])
    return ddivisorsum

divisor_intersection_dd(q[0]*q[3])
```

Now we can define the LHS:

```sage
def LHS_from_monomial(poly):
    degree = poly.degree()
    deltadeg = degree-1
    #start with the degree 0 thing for <delta1,deltan-1,delta>_0
    LHS=S(a_name(partition_from_poly(poly)))*divisor_intersection_d(poly)*divisor_intersection_dd(poly)

    
    for submonomial in R.monomial_all_divisors(poly):
        if submonomial != poly:
            partb = partition_from_poly(R(submonomial))
            parta = partition_from_poly(R.monomial_quotient(R(poly), R(submonomial)))
            quot = R.monomial_quotient(R(poly), R(submonomial))

            LHS+=k[0]*S(b_name(deltadeg,partb))*S(a_name(parta))*divisor_intersection_d(quot)*divisor_intersection_dd(quot)

    return LHS

LHS_from_monomial(q[0]*q[1]*q[2]).coefficient({d[0]: 1, dd[0]:1})


```

```sage
def eqns_from_monomial(poly):
    eqns=[]
    kLHS =k[0]*LHS_from_monomial(poly)
    kRHS = kRHS_from_monomial(poly)
    for i in range(n):
        for j in range(n):
            if (kLHS-kRHS).coefficient({d[i]: 1, dd[j]:1}) !=0:
                eqns.append((kLHS-kRHS).coefficient({d[i]: 1, dd[j]:1})      )
    return eqns


from IPython.display import display, Math
display(Math(latex(eqns_from_monomial(q[0]*q[1]))))
```

Add the starting equations:

```sage
trivialeqn=[]
for part in Partitions(1):
    trivialeqn.append(S(a_name(Partitions(1)[0]))-1)

trivialeqn
```

```sage
len(Partitions(3)[2])

#degree d representative polynomials
def degdpolys(d):
    all=[]
    for partition in Partitions(d):
        mon=1
        for l in range(len(partition)):
            mon*=q[l]^partition[l]
        all.append(mon)
    return all

degdpolys(4)

#eqns in degree m
def eqns(m):
    total=[]
    for p in degdpolys(m):
        total+=eqns_from_monomial(p)
    return total

eqns(3)
```

```sage
a_vars=var(a_paramnames)
b_vars=var(b_paramnames)
unknowns = a_vars+b_vars
tosolve = [SR(f) == 0 for f in eqns(2)+eqns(3)+eqns(4)+eqns(5)+eqns(6)+trivialeqn]
solutions = solve(tosolve, unknowns, solution_dict=True)

display(Math(latex(solutions)))
```
